```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier as dtree #import algorytmu przewidujacego
from sklearn.model_selection import train_test_split #Dzielenie na test i train
from sklearn.metrics import accuracy_score #sprawdzanie dokladnosci
music_data = pd.read_csv('music.csv')
X = music_data.drop(columns=['genre']) #Tutaj wyznaczamy, jakie kolumny chcemy zeby sie nie pokazywaly
y = music_data['genre'] # It returns all the values in the given column
# Zeby sprawdzic powyzsze po prostu print x lub y
# Nastepnie musimy przygotowac model, uzywajac algorytmu. W tym przypadku bedzie to alg "DecisionTree", czyli import
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = dtree()
model.fit(X_train, y_train)

# Teraz np chcielibysmy przewidziec jaka muzyke moze sluchac 21 letni mezczyzna i 22 letnia kobieta
predictions = model.predict(X_test) #W tym srodkowym, kwadratowym nawiasie podajemy wlasnie 21yo male i 22yo female do naszych danych, a w teorii on powinien lubic hipHop, a ona dance
#predictions  i tutaj nam podaje, ze dla 21yo male to bedzie dance... and so on

#Czesc druga. Dokladnosc
# Nie zawsze bedzie tak latwo dlatego czasem trzeba troche przekalkulowac dokladnosc takiego modelu

 #w tym x_train, x_test, y_train and so on... Tutaj wyjdzie nam tupple. A dla testowania podajmy 20% (0.2) naszych danych
# i zmieniamy model.fit dodajac tam _tran do x i y, i w predictions zamiast podawac jakies 2 kwadratowe nawiasy
# predictions = model.predict([ [21,1], [22, 0] ]) podajemy tylko X_train
#Aby sprawdzic dokladnosc takiego predictions musimy porownac z wartosciami jakie mamy w naszych wartosciach wyjsciowych dla "testing"
score = accuracy_score(y_test, predictions)
score
```

```python
# Cwiczenie 4. Wizualizacja
import pandas as pd
from sklearn.tree import DecisionTreeClassifier as dtree
from sklearn import tree

music_data = pd.read_csv('music.csv')
X = music_data.drop(columns=['genre']) # W tych 3 linijkach importujemy date
y = music_data['genre']

model = dtree() # Tworzymy model
model.fit(X, y) #Uczymy go

tree.export_graphviz(model, out_file='music-recommender.dot',
                    feature_names=['age', 'gender'],
                    class_names = sorted(y.unique()),
                    label='all',
                    rounded = True,
                    filled = True)
```
